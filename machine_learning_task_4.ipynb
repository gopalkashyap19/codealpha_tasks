```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
def clean_data(df):
    print("\n🔹 Cleaning Dataset...")
    print("Original Shape:", df.shape)


    df.dropna(axis=1, how='all', inplace=True)


    df.drop_duplicates(inplace=True)


    print("\nMissing Values Before:\n", df.isnull().sum())


    for col in df.columns:
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            df[col].fillna(df[col].median(), inplace=True)

    print("\nMissing Values After:\n", df.isnull().sum())
    print("Cleaned Shape:", df.shape)

    return df
```

```python
heart_df = pd.read_csv("heart.csv")
heart_df = clean_data(heart_df)

X_heart = heart_df.drop("target", axis=1)
y_heart = heart_df["target"]

scaler = StandardScaler()
X_heart = scaler.fit_transform(X_heart)

X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(
    X_heart, y_heart, test_size=0.2, random_state=42)
```

```python
diab_df = pd.read_csv("diabetes.csv")
diab_df = clean_data(diab_df)

X_diab = diab_df.drop("Outcome", axis=1)
y_diab = diab_df["Outcome"]

scaler = StandardScaler()
X_diab = scaler.fit_transform(X_diab)

X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(
    X_diab, y_diab, test_size=0.2, random_state=42)
```

```python
cancer_df = pd.read_csv("brest_cancer.csv")
cancer_df = clean_data(cancer_df)

if 'id' in cancer_df.columns:
    cancer_df.drop('id', axis=1, inplace=True)

cancer_df['diagnosis'] = cancer_df['diagnosis'].map({'M':1, 'B':0})

X_cancer = cancer_df.drop("diagnosis", axis=1)
y_cancer = cancer_df["diagnosis"]

scaler = StandardScaler()
X_cancer = scaler.fit_transform(X_cancer)

X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X_cancer, y_cancer, test_size=0.2, random_state=42)
```

```python
def train_models(X_train, X_test, y_train, y_test, disease_name):
    models = {
        "Logistic Regression": LogisticRegression(max_iter=1000),
        "SVM": SVC(),
        "Random Forest": RandomForestClassifier(),
        "XGBoost": XGBClassifier()
    }

    print(f"\n{'='*50}")
    print(f"🏥 {disease_name} Prediction Results")
    print(f"{'='*50}")

    best_acc = 0
    best_model = None

    for name, model in models.items():
        model.fit(X_train, y_train)
        preds = model.predict(X_test)
        acc = accuracy_score(y_test, preds)

        print(f"\n{name}")
        print("Accuracy:", round(acc,4))
        print(classification_report(y_test, preds))

        if acc > best_acc:
            best_acc = acc
            best_model = model

    return best_model
```

```python
best_heart_model = train_models(X_train_h, X_test_h, y_train_h, y_test_h, "Heart Disease")
best_diab_model = train_models(X_train_d, X_test_d, y_train_d, y_test_d, "Diabetes")
best_cancer_model = train_models(X_train_c, X_test_c, y_train_c, y_test_c, "Breast Cancer")
```

```python
joblib.dump(best_heart_model, "heart_disease_model.pkl")
joblib.dump(best_diab_model, "diabetes_model.pkl")
joblib.dump(best_cancer_model, "brest_cancer_model.pkl")

print("\n✅ All models saved successfully!")
```

```python
preds = best_heart_model.predict(X_test_h)
cm = confusion_matrix(y_test_h, preds)

sns.heatmap(cm, annot=True, fmt='d')
plt.title("Heart Disease Confusion Matrix")
plt.show()
```

```python
preds_diab = best_diab_model.predict(X_test_d)
cm_diab = confusion_matrix(y_test_d, preds_diab)

plt.figure()
sns.heatmap(cm_diab, annot=True, fmt='d')
plt.title("Diabetes Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
preds_cancer = best_cancer_model.predict(X_test_c)
cm_cancer = confusion_matrix(y_test_c, preds_cancer)

plt.figure()
sns.heatmap(cm_cancer, annot=True, fmt='d')
plt.title("Brest Cancer Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
